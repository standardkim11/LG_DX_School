```python
kiwi.add_user_word("딥시크", "NNP")
```

```python
def pos_tagging(string):
    string = kiwi.space(string)
    pos_words = kiwi.tokenize(string, stopwords = stopwords)
    result = [i[0] for i in pos_words if i[1] in {'NNG','NNP'}]
    return result
```

```python
tagged_pos_result = []

for i in text:
    tagged_pos_result.append(pos_tagging(i))
```

## 2. 빈도계산하기

```python
from collections import Counter
```

```python
result = []

for doc in tagged_pos_result:
    for n in doc:
        result.append(n)
```

```python
count = Counter(result)
count
```

## 3. 트리맵그리기

```python
#pip install squarify
```

```python
import squarify
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
import matplotlib.font_manager as fm
```

```python
top_15 = count.most_common(15)
```

```python
label = [i[0] for i in top_15 ] # key
sizes = [i[1]for i in top_15] # value
```

```python
norm = matplotlib.colors.Normalize(vmin = min(sizes), vmax = max(sizes)) #정규화
```

```python
colors = [matplotlib.cm.Blues(norm(v)) for v in sizes] 
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['font.size'] = 13
```

```python
plt.figure(figsize = (8,6))
squarify.plot(sizes = sizes , label = label  , color = colors , bar_kwargs = dict(linewidth = 5 , edgecolor = '#eee'))
plt.title('트리맵')
plt.axis('off')
plt.show()
```
