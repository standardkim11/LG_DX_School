```python
import nltk
nltk.download("punkt")
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('stopwords')
nltk.download('tagsets_json')
```

```python
import pandas as pd

df = pd.read_csv('/content/sample_data/ÏòÅÏñ¥ ÌÅ¨Î°§ÎßÅ.csv')
```

```python
df.head()
```

```python
stopwords = nltk.corpus.stopwords.words('english')
def pos_tagging(string):
    string = string.lower()
    string = nltk.word_tokenize(string)
    string = nltk.pos_tag(string)

    result = [w for w, t in string if t in ['NN','NNS','NNP','NNPS'] and w not in stopwords]
    return result
```

```python
from tqdm import tqdm
tagged_pos_document = []

for i in tqdm(df['Review']):
    tagged_pos_document.append(pos_tagging(i))
```

## 2. Doc2Vec
- vector_size: Î¨∏ÏÑú Î≤°ÌÑ∞Ïùò Ï∞®Ïõê
- alpha : Î™®Îç∏ ÌïôÏäµÏãú Ï¥àÍ∏∞ÌïôÏäµÎ•†
- min_alpha: alphaÍ∞Ä Ïù¥ Í∞í Ïù¥ÌïòÎ°ú Í∞êÏÜåÌïòÏßÄ ÏïäÎèÑÎ°ù Ï†úÌïúÌïòÎäî Í∞í
- window: Ï£ºÎ≥Ä Îã®Ïñ¥Î•º Í≥†Î†§ÌïòÎäî Î≤îÏúÑ
- min_count: Îã®Ïñ¥ ÏµúÏÜå Îì±Ïû• ÌöüÏàò
- epochs: ÌïôÏäµ Î∞òÎ≥µ ÌöüÏàò
- dm: 1ÏùÄ PV-DM, 0ÏùÄ PV-DBOW

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
tagged_corpus_list = []

for i, j in enumerate(tagged_pos_document):
    tag = 'document_{}'.format(i)
    tagged_corpus_list.append(TaggedDocument(words = j, tags = [tag]))
```

```python
tagged_corpus_list[0]
```

```python
#Î™®Îç∏ÏÑ†Ïñ∏

doc2vec_model = Doc2Vec(vector_size = 300, alpha = 0.025, min_alpha = 0.01, window = 4, min_count = 1, epochs= 100, dm = 0)
```

```python
#Îã®Ïñ¥ÏÇ¨Ï†Ñ
doc2vec_model.build_vocab(tagged_corpus_list)
```

```python
#ÌïôÏäµ

doc2vec_model.train(tagged_corpus_list, total_examples = doc2vec_model.corpus_count, epochs = doc2vec_model.epochs)
```

```python
#Î¨∏ÏÑúÍ∞Ñ Ïú†ÏÇ¨ÎèÑ

doc2vec_model.dv.most_similar('document_0',topn = 20)
```

```python
#ÏÉàÎ°úÏö¥ Î¨∏ÏÑúÏóê Î≤°ÌÑ∞Í∞í ÎßåÎì§Í∏∞
new_doc_vec = doc2vec_model.infer_vector(['lg','app','is','good'])
```
