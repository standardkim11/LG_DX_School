## 1. 정규표현식

```python
import re
```

```python
pattern = re.compile('패턴')
re.search(pattern,'text')
re.findall(pattern,'text')
re.sub(pattern, '대치할 문자','text')
```

### 1.1 전화번호

```python
#search - 서울전화 뽑기
text = '사무실전화는요 02-453-1234 이구요 핸드폰번호는 010-2222-4444 입니다.'

pattern = re.compile(r'02-\d{3}-\d{4}')
re.search(pattern, text)[0]
```

```python
#findall - 서울전화 뽑기
text = '사무실전화는요 02-453-1234 이구요, 02-333-6666 핸드폰번호는 010-2222-4444 입니다.'

pattern = re.compile(r'\d{2}-\d{3}-\d{4}')
re.findall(pattern, text)
```

```python
#findall - 핸드폰번호 뽑기
text = '사무실전화는요 02-453-1234 이구요, 02-333-6666 핸드폰번호는 010-2222-4444 입니다. 010-3333-2222'

pattern = re.compile(r'\d{3}-\d{4}-\d{4}')
re.findall(pattern, text)
```

##### ♦️1. 실습예제
  - 유선번호 핸드폰 번호 전체 뽑기

```python
text = '사무실전화는요 02-453-1234 이구요, 02.333.6666 핸드폰번호는 010-2222-4444 입니다. 010.3333.2222'

pattern = re.compile(r'\d{2,3}\W\d{3,4}\W\d{4}')
re.findall(pattern, text)
```

### 1.2 한글 자음 제거 - re.sub

```python
#출력하기 - Search
text = '안녕하세요ㅎㅎㅎㅎㅎ ㅋㅋㅋㅋㅋ'
pattern = re.compile(r'[ㄱ-ㅎ]+')
re.search(pattern,text)[0]
```

```python
#출력하기 - findall
pattern = re.compile(r'[ㄱ-ㅎ]+')
re.findall(pattern,text)
```

```python
#제거하기 - resub
pattern = re.compile(r'[ㄱ-ㅎ]+')
re.sub(pattern, '*' , text)
```

##### ♦️2. 실습예제
  - 알파벳 전체 필터링해서 *** 문자로 대체하기
  - 예상 결과 : '*** 반갑습니다. 나는 *** 입니다.'

```python
text = 'Hello 반갑습니다. 나는 Olivia 입니다.'
pattern = re.compile(r'[A-Za-z]+')
re.sub(pattern,'***',text)
```

##### ♦️2.1. 번외 실습예제 : ^ 사용해보기
  - 알파벳 전체 필터링해서 *** 문자로 대체하기
  - 예상 결과 : '*** 반갑습니다. 나는 *** 입니다.'

```python
pattern = re.compile(r'[^\s가-힣\.]+')
re.sub(pattern,'***',text)
```

### 1.4 특수기호

```python
text = 'Hello 반갑습니다. ~~~!!! 나는 Olivia 입니다.~~~!!!'
pattern = re.compile(r'[A-Za-z\s가-힣\.\s]+')
re.sub(pattern,'',text)
```
