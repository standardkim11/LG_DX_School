## 1. TF-IDF : 형태소나누기

```python
sample_text = ["데이터 과학은 과학적인 방법, 과정, 알고리즘, 시스템을 사용하여 구조화된 데이터와 비구조화된 데이터에서 지식과 통찰을 추출하는 학문 분야입니다.",
               "머신 러닝은 인공지능의 한 분야로, 알고리즘과 통계적 모델을 사용하여 컴퓨터가 경험을 통해 작업 성능을 개선하도록 하는 기술입니다.",
               "인공지능은 인간의 지능을 기계에서 시뮬레이션하는 것으로, 인간처럼 생각하고 행동하도록 프로그램된 시스템을 말합니다."]
```

```python
sample_text
```

```python
from kiwipiepy import Kiwi
kiwi = Kiwi()
```

```python
documents = []

for doc in sample_text:
    doc_pos = [i[0] for i in kiwi.tokenize(doc)]
    documents.append(' '.join(doc_pos))
```

## 3. 워드클라우드로 시각화

```python
from wordcloud import WordCloud
```

```python
from kiwipiepy.utils import Stopwords
```

```python
stopwords = Stopwords()
```

```python
def pos_tagging(string):
    string = kiwi.space(string)
    pos_words = kiwi.tokenize(string, stopwords = stopwords)
    result = [i[0] for i in pos_words if i[1] in {'NNG','NNP'}]
    return result
```

```python
tagged_pos_result = []

for i in sample_text:
    tagged_pos_result.append(pos_tagging(i))
```

```python
from collections import Counter
```

```python
result = []

for doc in tagged_pos_result:
    for n in doc:
        result.append(n)
```

```python
count = Counter(result)
count
```

```python
import matplotlib.pyplot as plt
```

```python
wc = WordCloud(font_path = "C:\Windows\Fonts\malgun.ttf",
               height = 400, width = 400,
               background_color = 'white').generate_from_frequencies(count)
plt.figure(figsize = (10,8))
plt.imshow(wc , interpolation = 'bilinear') #이미지를 부드럽게
plt.axis('off')
plt.show()
```

```python
freqs = tfidf_df.max(axis=1).to_dict()

wc = WordCloud(
    font_path="C:/Windows/Fonts/malgun.ttf",  # 한글 폰트 경로
    width=1000, height=600, background_color="white"
).generate_from_frequencies(freqs)

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
