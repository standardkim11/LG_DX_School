```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast
```

```python
df = pd.read_excel('./Data/영어크롤링임베딩.xlsx.xlsx', index_col = 0)
df.head()
```

```python
df.isnull().sum()
```

```python
# array형태를 숫자형태로 변환

def erase_double_blank(string):
    result = string
    while "  " in result:
        result = result.replace("  ", " ")
    while ",," in result:
        result = result.replace(",,", ",")
    while ", ," in result:
        result = result.replace(", ,", ",")
    result = result.replace("[, ", "[")
    result = result.replace(", ]", "]")
    return result

df["vector"] = df["vector"].apply(lambda x : x.replace("\n", " "))
df["vector"] = df["vector"].apply(lambda x : x.replace(" ", ", "))
df["vector"] = df["vector"].apply(lambda x : erase_double_blank(x))
 
df["vector"] = df["vector"].apply(lambda x : ast.literal_eval(x))
```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
k = 3

model = KMeans(n_clusters = k)
```

```python
arr_vec = np.array(list(df['vector']))
```

```python
model.fit(arr_vec)
```

```python
df['kmeans'] = model.fit_predict(arr_vec)
```

```python
df.head()
```

```python
# DBSCAN

from sklearn.cluster import DBSCAN
```

```python
model_dbscan = DBSCAN(eps = 0.9, min_samples = 5)

model_dbscan.fit(arr_vec)
```

```python
df['dbscan'] = model_dbscan.fit_predict(arr_vec)
```

```python
df.head()
```

```python
df['dbscan'].unique()
```

```python
# 계층적 군집분석

from sklearn.cluster import AgglomerativeClustering
```

```python
model_agg = AgglomerativeClustering(n_clusters= 3)
```

```python
model_agg.fit(arr_vec)
```

```python
df['agg'] = model_agg.fit_predict(arr_vec)
```

```python
df.head()
```

```python
from sklearn.metrics import silhouette_score
```

```python
print(silhouette_score(arr_vec, df['kmeans'] , metric = 'euclidean'))
```

```python
print(silhouette_score(arr_vec, df['dbscan'] , metric = 'euclidean'))
```

```python
print(silhouette_score(arr_vec, df['agg'] , metric = 'euclidean'))
```

```python
from sklearn.decomposition import PCA
```

```python
#차원축소 (주성분)
pca = PCA(n_components = 2)
principalComponents = pca.fit_transform(arr_vec)
```

```python
principalComponents
```

```python
df_pca = pd.DataFrame(principalComponents , columns = ['pca1','pca2'])
```

```python
df_pca.head()
```

```python
df_pca[['kmeans', 'dbscan','agg']] = df[['kmeans', 'dbscan','agg']]
```

```python
plt.figure(figsize = (10,7))

plt.scatter(df_pca['pca1'], df_pca['pca2'])
plt.show()
```

# **와인 데이터셋**

```python
wine= pd.read_csv('./Data/wine.csv.csv', index_col = 0)
```

```python
wine = wine.iloc[:,2:-1]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scale_df  = pd.DataFrame(scaler.fit_transform(wine), columns = wine.columns)
```

```python
md_db_wine = DBSCAN(eps = 0.9)
```

```python
md_db_wine.fit(scale_df)
```

```python
scale_df['dbscan'] = md_db_wine.fit_predict(scale_df)
```

```python
scale_df['dbscan'].unique()
```

```python
md_km_wine = KMeans(n_clusters=2)
```

```python
md_km_wine.fit(scale_df)
```

```python
scale_df['kmeans'] = md_km_wine.fit_predict(scale_df)
```

```python
wine_pca = PCA(n_components= 3)
pca_result = wine_pca.fit_transform(scale_df)
```

```python
wine_pca = pd.DataFrame(pca_result, columns = ['pca1','pca2','pca3'])
```

```python
wine_pca['cluster'] = pd.DataFrame(scale_df['dbscan'])
```

```python
wine_pca['kcluster'] = pd.DataFrame(scale_df['kmeans'])
```

```python
wine_pca
```

```python
from mpl_toolkits.mplot3d import Axes3D
# 3D 시각화 함수
def plot_clusters_3d(df, label_col, title):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    sc = ax.scatter(df['pca1'], df['pca2'], df['pca3'], 
                    c=df[label_col], cmap='tab10', s=50, alpha=0.8)
    ax.set_title(title)
    ax.set_xlabel("PCA 1")
    ax.set_ylabel("PCA 2")
    ax.set_zlabel("PCA 3")
    #fig.colorbar(sc, ax=ax, label="Cluster")
    plt.tight_layout()
    plt.show()
```

```python
plot_clusters_3d(wine_pca, 'cluster' ,'pca')
```

```python
plot_clusters_3d(wine_pca, 'kcluster' ,'pca')
```
