```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split ,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import root_mean_squared_error , mean_squared_error , accuracy_score, r2_score, confusion_matrix , classification_report ,f1_score
from sklearn.linear_model import LogisticRegression , LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC , SVR
from sklearn.neighbors import KNeighborsClassifier , KNeighborsRegressor
from sklearn.ensemble import RandomForestClassifier , GradientBoostingClassifier
from xgboost import XGBClassifier

from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv('./Data/bmi_500.csv')
```

# **실습**

```python
df = pd.read_csv('./Data/sample_data.csv', index_col = 0)
```

```python
df.head(5)
```

```python
df.isna().sum()
```

```python
df = df.drop(columns = ['cst_id_di' , 'cst_id_di.1'])
```

```python
df['MRC_ID_DI'].value_counts().sort_values()
```

```python
for col in df.columns:
    if df[col].nunique() <= 12: 
        print(col, df[col].unique())
```

```python
cat_cols = [col for col in df.columns if df[col].nunique() <= 12]
```

```python
num_cols = []
for col in df.columns:
    if col not in cat_cols:
        num_cols.append(col)
```

```python
MRC_ID_DI = df['MRC_ID_DI'].value_counts()
others = MRC_ID_DI[MRC_ID_DI <= 150].index
df['MRC_ID_DI_bin'] = df['MRC_ID_DI'].replace(others, 999)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['MRC_ID_DI_bin'] = le.fit_transform(df['MRC_ID_DI_bin'])
```

```python
df['MRC_ID_DI_bin'].value_counts().sort_values()
```

```python
x = df.drop(columns = ['MRC_ID_DI_bin', 'MRC_ID_DI'], axis= 1)
y = df['MRC_ID_DI_bin']
y_original = df['MRC_ID_DI']
```

```python
X_train, X_test, y_train, y_test, y_original_train, y_original_test = train_test_split(
    x, y, y_original,
    test_size=0.2,
    random_state=42
)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
smote = SMOTE(random_state=42, k_neighbors=2)
X_train_res, y_train_res = smote.fit_resample(X_train_scaled, y_train)
```

```python
md_rf = RandomForestClassifier(random_state = 42, class_weight='balanced')
```

```python
md_rf.fit(X_train_res, y_train_res)
```

```python
pred_rf = md_rf.predict(X_test_scaled)
```

```python
f1_score(y_test,pred_rf , average = 'macro')
```

```python
cm_rf = confusion_matrix(y_test, pred_rf)
print(classification_report(y_test, pred_rf))
 
plt.figure(figsize=(4, 4))
sns.heatmap(cm_rf, annot=True, fmt="d", cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
model_xgb = XGBClassifier(random_state=42)
```

```python
model_xgb.fit(X_train_res, y_train_res)
```

```python
pred_xgb = model_xgb.predict(X_test_scaled)
```

```python
f1_score(y_test,pred_xgb , average = 'macro')
```

```python
n_estimators = [100, 300, 500, 800, 1000]
max_depth = [5, 10, 15, 20, 25, 30]

parameter_grid = {
    'n_estimators': n_estimators,
    'max_depth': max_depth
}
```

```python
grid_cv = GridSearchCV(estimator=model_xgb,
                               param_grid  = parameter_grid,
                               cv=5,
                               scoring='f1_macro',  
                               n_jobs=-1,
                               verbose=1,
                               return_train_score=True
                      )
```

```python
grid_cv.fit(X_train_res, y_train_res)
```

```python
best_params = grid_cv.best_params_
best_score = grid_cv.best_score_
print(f'Best parameters: {best_params}')
print(f'Best score: {best_score}')
```

```python
best_model_xgb = grid_cv.best_estimator_
pred_xgb2 = best_model_xgb.predict(X_test_scaled)
```

```python
f1_score(y_test,pred_xgb2 , average = 'macro')
```

```python
accuracy2 = accuracy_score(y_test, pred_xgb2)
cm2 = confusion_matrix(y_test, pred_xgb2)
plt.figure(figsize=(4, 4))
sns.heatmap(cm2, annot=True, fmt="d", cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

**0으로 분류한 데이터는 올바르게 분류한 데이터 이므로 제외하고**   
**1(합친 데이터)를 풀어서 다시 모델 제작후 학습 시키는 계층적 분류를 사용한다.**
**그리고 2차분석에서 사용할 train set를 미리 가져와야함**

```python
# 전체 테스트셋에 대한 이진 예측
pred_xgb2 = best_model_xgb.predict(X_test_scaled)

# 2차 분석에 사용할 데이터

mask_test_idx = np.where(pred_xgb2 == 1)[0]
X_2_test = X_test.iloc[mask_test_idx].copy()
y_2_test = y_original_test.iloc[mask_test_idx].copy()

# 2차 분석 
X_2_test_scaled = scaler.transform(X_2_test) # 기존 스케일러 사용
xgb_model2 = XGBClassifier(**best_params, random_state=42)

y_2_train_enc = y_2_train - 1
xgb_model2.fit(X_2_train_scaled, y_2_train_enc)

pred_xgb2_enc = xgb_model2.predict(X_2_test_scaled)
pred_xgb3 = pred_xgb2_enc + 1

# 최종 예측
y_final_pred = pred_xgb2.copy()

# 1차 예측에서 1로 분류된 위치에 2차 예측값을 대입
y_final_pred[mask_test_idx] = pred_xgb3

# 최종 성능 평가
print("계층적 최종 F1 Score:", f1_score(y_original_test, y_final_pred, average='weighted'))
print(classification_report(y_original_test, y_final_pred))
```

```python
n_estimators = [100, 300, 500, 800, 1000]
max_depth = [5, 10, 15, 20, 25, 30]
learning_rate = [0.01, 0.05, 0.1]
parameter_grid = {
    'n_estimators': n_estimators,
    'max_depth': max_depth,
    'learning_rate' : learning_rate
}
```

```python
from sklearn.model_selection import RandomizedSearchCV


random_search = RandomizedSearchCV(
    model_xgb, param_distributions=parameter_grid,
    n_iter=10, cv=3, scoring='f1_macro', n_jobs=-1, verbose=1
)
random_search.fit(X_2_train_scaled, y_2_train_enc)
```

```python
best_params2 = random_search.best_params_
best_score2 = random_search.best_score_
print(f'Best parameters: {best_params2}')
print(f'Best score: {best_score2}')
```

```python
final_model_xgb = random_search.best_estimator_
final_pred_xgb = final_model_xgb.predict(X_2_test_scaled)
```

```python
y_final_pred = pred_xgb2.copy()

final_pred_xgb_mapped = final_pred_xgb + 1 
y_final_pred[mask_test_idx] = final_pred_xgb_mapped

# 최종 성능 평가
print("계층적 최종 F1 Score:", f1_score(y_original_test, y_final_pred, average='weighted'))
print(classification_report(y_original_test, y_final_pred))
```
