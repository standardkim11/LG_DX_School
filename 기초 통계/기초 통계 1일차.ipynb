# **기초 통계**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Data/bmi_500.csv')
df.head()
```

```python
np.mean(df['Height'])
```

```python
np.var(df['Height'])
```

```python
np.var(df['Height'], ddof =1 ) # ddof = 1 : 표본집단의 분산을 표현하고 n-1로 보정해서 과소 평가를 막음. 분산이 커짐  
```

```python
np.std(df['Height'])
```

```python
np.std(df['Height'], ddof = 1) #표본 표준편차 보정 
```

```python
plt.hist(df['Height'])
```

```python
# 표준

m = np.mean(df['Height'])
s = np.std(df['Height'])

normalized = [(i-m)/s for i in df['Height']]
```

```python
plt.hist(normalized)
```
